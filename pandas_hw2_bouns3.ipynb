```python
# считываем первую таблицу рейтинги
data = pd.read_csv('C:\\Users\\alexander.dubovik\\Documents\\PYDA-38\\NUMP_PAND\\hw\\pandas_hw1\\ml-25m\\ratings.csv')
data.head()
```

```python
def timestamp_to_date (row):
    '''
    Из формата времени Unix переводим в стандартное
    '''
    return datetime.date.fromtimestamp(row['timestamp'])
```

```python
%%time
# дополнил условие (перевести дату ввода рейтинга из UTC в стандартный формат) оно не пригодилось, 
# но возможно захочется еще сделать сортировку, в каком году (месяце) за какой фильм голосовали больше

data['timest'] = data.apply(timestamp_to_date, axis = 1)
```

```python
# считываем первую таблицу фильмы
movies = pd.read_csv('C:\\Users\\alexander.dubovik\\Documents\\PYDA-38\\NUMP_PAND\\hw\\pandas_hw1\\ml-25m\\movies.csv')
movies.head()
```

```python
import re
```

```python
def production_year (row = 1900):
    '''
    Выделяем дату из колонки с названем фильма, при условию, что дата должно быть в скобках, исключая вывода даты из названия самого фильма. Если года нет, то подставляем 1900
    '''
    date_ = ''.join(re.findall(r'[(](\d{4})[)]', row['title'], flags=re.IGNORECASE)).strip()
    if date_:
        return int(float(date_))
```

```python
# применяем функцию к фильмам, добавляем колонку с годами
movies['years'] = movies.apply(production_year, axis=1)
movies.head()
```

```python
# применяем фильтр и находим все фильмы, которые подпадают под условие: список из всех годов с 1950 по 2010 года
movies_filtered = movies[(movies['years'] >= 1950) & (movies['years'] <= 2010)]
movies_filtered.head()
```

```python
%%time
# объединяем два ДФ в один, по movieID. Получаем новый ДФ с названиями фильмов, рейтингов и iduser

movie_merge = pd.merge(movies_filtered, data, how='left', on='movieId')
movie_merge.head()
```

```python
# Делаем группировку по году и названию фильма, далее сортируем по году и по рейтингу. Я отсортировал в обратном порядке по обоим, так как не было условия как сортировать год

movie_merge.groupby(['years', 'title']).agg({'rating': 'mean'}).sort_values(by = ['years', 'rating'], ascending=[True, True]).head()
```
