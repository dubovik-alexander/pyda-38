```python
#перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

# перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def get_name_using_number(document, num_doc):
    '''
    Функция нахождения имени в документах по номеру
    '''
    #num_doc = input('Введите номер')
    for num in document:
        if num_doc == num['number']:
            pers_name = (f"Владелец документа: {num['name']}")
            break
        else:
            pers_name = 'Документ не найден в базе'
    return pers_name
```

```python
def get_shelf_using_number_first(shelfs, num_doc):
    for key, value in shelfs.items(): #разбиваем полки
        if num_doc in value:
            key_num = key
            break
        else:
            key_num = False
    return key_num
```

```python
def get_shelf_using_number(shelfs, num_doc):
    '''
    функция нахождения на какой полке лежит документ по номеру
    '''
    #num_doc = str(input('Введите номер'))
    key_num = get_shelf_using_number_first(shelfs, num_doc)
    if  key_num == False:
        shelf_num = 'Документ не найден в базе'
    else:
        shelf_num = (f'Документ хранится на полке: {key_num}')
    return shelf_num
```

```python
def get_all_data(document, shelf_num):
    '''
    Функция вывода полной информации по всем документам
    '''
    for num in document: # перебираем все данные из документов
        for key, value in shelf_num.items(): # берем полку и нам нужен отсюда ключ
            for num_el in value: # беру номер документа и в if ищу его в документах, если есть, то печатаю уже все данные
                if num['number'] == num_el:#    doc_data = (f"№: {num['number']}, тип: {num['type']}, владелец: {get_name_using_number(document)}, полка хранения: {get_shelf_using_number(shelf_num)}")
                    print(f"№: {num['number']}, тип: {num['type']}, владелец: {num['name']}, полка хранения: {key}")
                
    return 
```

```python
def check_shelf (shelfs, num_doc):
    '''
    Функция проверки полок на наличие
    '''
    if num_doc in shelfs:
        x = True
    else: 
        x = False
    return (x)
```

```python
def add_shelf(shelfs, num_doc):
    '''
    функция добавления полок
    '''
    #num_doc = str(input('Введите номер'))
    if check_shelf (shelfs, num_doc) == True:
    ###if num_doc in shelfs:
        print(f'Такая полка уже существует. Текущий перечень полок: {[key for key in shelfs.keys()]}')
    else:
        shelfs[num_doc] = []
        print(f'Полка добавлена. Текущий перечень полок: {[key for key in shelfs.keys()]}')
    return

```

```python
def delete_shelf_if_empty(shelfs, num_doc):
    '''
    функция удаление полок, если выбранная полка пустая
    '''
    #num_doc = str(input('Введите номер'))
    if check_shelf (shelfs, num_doc) == "True": # есть ли такая полка
        if len(directories.get(num_doc)) == 0: # проверяем наличие values через длину
            del(shelfs[num_doc])
            print(f'Полка удалена. Текущий перечень полок: {[key for key in shelfs.keys()]}')
        else:
            print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {[key for key in shelfs.keys()]}')
    else:
        print(f'Такой полки не существует. Текущий перечень полок: {[key for key in shelfs.keys()]}')
    return
```

```python
def add_new_document(documentation, shelfs):
    '''
    функция добавления новых документов
    '''
    num_doc = str(input('Введите номер документа: '))
    tp_doc = str(input('Введите тип документа: '))
    pers_doc = str(input('Введите владельца документа: '))
    shelf_doc = str(input('Введите полку для хранения: '))
    if check_shelf (shelfs, shelf_doc) == 'False': # проверяю есть ли такая папка
        print('Такой полки не существует. Добавьте полку командой as. \n Текущий список документов:') 
        #get_all_data(documents, directories)
    else:
        shelfs[shelf_doc].append(num_doc) # вставляем данные в полки
        documentation.append(dict(zip(['type', 'number', 'name'], [tp_doc, num_doc, pers_doc]))) # обновляем документы
        print(f'Документ добавлен. Текущий список документов:')
    get_all_data(documents, directories)
    print('')
    return

```

```python
# застрял нужна помощь

def delete_document(document, shelfs):
    '''
    функция удаление по номеру документа из полок и из списка документов
    '''
    num_doc = str(input('Введите номер документа: '))
    if check_shelf (shelfs, shelf_doc) == 'False': # проверяю есть ли такая папка
        print('Документ не найден в базе. Текущий список документов:') 
        #get_all_data(documents, directories)
    else:
        key_shelfs = str(get_shelf_using_number_first(shelfs, num_doc)) # получаем номер ключа полки, на которой лежит документ
        shelfs[key_shelfs].remove(num_doc)    
        for num in documents: # перебираем все данные из документов
            if num['number'] == num_doc:
                del num['number'], num['type'], num['name']
                break
        print(f'Документ удален. Текущий список документов: ')
    
    get_all_data(documents, directories)
    return
```

```python
def main(documentation, shelf):
    """ 
   Главная функция, которая позволят по командам пользователя:
   "p" может узнать владельца документа по его номеру;
   "s" может по номеру документа узнать на какой полке он хранится; 
   "l" может увидеть полную информацию по всем документам;
   "ads" может добавить новую полку;
   "ds" может удалить существующую полку из данных (только если она пустая);
   
   "q" - выход
    """
    while True:
        # Задание 1:
        user_cmd = str(input('Введите команду'))
        if user_cmd == 'p':                         # Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру
            num_doc = str(input('Введите номер'))
            print(get_name_using_number(documentation, num_doc))
        elif user_cmd == 's':                      # Пункт 2. Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
            num_doc = str(input('Введите номер'))
            print(get_shelf_using_number(shelf, num_doc))
        elif user_cmd == 'l':                      # Пункт 3. Пользователь по команде "l" может увидеть полную информацию по всем документам
            get_all_data(documentation, shelf)
        elif user_cmd == 'ads':                    # Пункт 4. Пользователь по команде "ads" может добавить новую полку
            num_doc = str(input('Введите номер'))
            add_shelf(shelf, num_doc)
        elif user_cmd == 'ds':                     # Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
            num_doc = str(input('Введите номер'))
            delete_shelf_if_empty(shelf, num_doc)
        # Задание 2 (необязательное): 
        elif user_cmd == 'ad':                     # Пункт 1. Пользователь по команде “ad” может добавить новый документ в данные
            add_new_document(documentation, shelf)
        elif user_cmd == 'd':                      # Пункт 2. Пользователь по команде “d” может удалить документ из данных
            delete_document(documentation, shelf)
        elif user_cmd == 'q':
            print('До свидавния!')
            break

main(documents, directories)
```
